```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['prognosis']),np.array(df['prognosis']), test_size=0.3,random_state=121)
print(X_train)
```

```python
#pred=svm.predict(X_test)
lg=RandomForestClassifier()
lg.fit(X_train,y_train)
pred=lg.predict(X_test)
print(pred[:10])
print(y_test[:10])
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(confusion_matrix(y_test,pred),classification_report(y_test,pred))
train_pred=lg.predict(X_train)
print(accuracy_score(y_train,train_pred))
```
